```python
import pandas as pd
import numpy as np
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

pd.DataFrame(exam_data)
```

```python
df = pd.DataFrame(exam_data, index = labels)
```

```python
df.head(3)
```

```python
df.dropna()
```

```python
df[["name","score"]]
```

```python
dict1 = {"name" : "Suresh", "score": 15.5, "attempts": 1, "qualify": "yes"}

data2 = pd.DataFrame([dict1], index = ["k"])

pd.concat([df,data2])
```

```python
df = pd.concat([df,data2])
```

```python
df.drop("attempts" , axis = 1)
```

```python
success = lambda x: 1 if x>10 else 0
df['success'] = df['score'].apply(success)
df
```
